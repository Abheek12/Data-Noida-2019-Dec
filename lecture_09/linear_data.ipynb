```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_features=1, noise=4, random_state=42, bias=100)
```

```python
X.shape
```

```python
y.shape
```

```python
plt.scatter(X, y)
```

```python
bias = 100
```

```python
X, y, coef = make_regression(n_features=1, noise=10, coef=True, random_state=42, bias=bias)
```

```python
coef, bias
```

```python
x_min, x_max = X.min(), X.max()
```

```python
X_gen = np.linspace(x_min, x_max, 10)
```

```python
y_gen = X_gen * coef + bias
```

```python
plt.plot(X_gen, y_gen)

plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.coef_, model.intercept_
```

```python
plt.plot(X_gen, y_gen)

plt.scatter(X, y)

plt.plot(X_gen, X_gen * model.coef_ + model.intercept_)
```
