```python
import random

window = 3

freq = {}

for k in range(window, len(speech)):
    word = speech[k-window: k]
    
    ch = speech[k]
    
    if word not in freq:
        freq[word] = {}
        
    if ch not in freq[word]:
        freq[word][ch] = 0
        
    freq[word][ch] += 1
```

```python
seed = "कौन बोला? कौन बोला?"
```

```python
for i in range(500):

    cut = seed[-window:]

    fqs = freq[cut]

    top_fqs = sorted(fqs.items(), key=lambda x: x[1], reverse=True)[:3]

    next_ch = random.choice(top_fqs)[0]

    seed = seed + next_ch
```
