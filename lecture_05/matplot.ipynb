```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.linspace(-10, 10, 10)
```

```python
y = X ** 2
z = X ** 3
```

```python
plt.figure()
plt.plot(X, y)
plt.show()

plt.figure()
plt.plot(X, z)
plt.show()
```

```python
plt.figure()
plt.plot(X, y, label="square")
plt.plot(X, z, label="cube")
plt.xlim(-10, 10)
plt.ylim(-100, 100)
plt.xlabel("This is for input")
plt.ylabel("This is for output")
plt.title("Poore 20 number")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.scatter(X, y, label="square", color="red", marker="^")
plt.scatter(X, z, label="cube", color="yellow")
plt.xlim(-10, 10)
plt.ylim(-100, 100)
plt.xlabel("This is for input")
plt.ylabel("This is for output")
plt.title("Poore 20 number")
plt.legend()
plt.show()
```

```python
ang = np.linspace(0, 20*np.pi, 1000)
```

```python
rad = np.linspace(0, 20, 1000)
```

```python
X = rad*np.cos(ang)
```

```python
y = rad*np.sin(ang)
```

```python
plt.plot(X, y)
```

```python
X = np.linspace(0, 2*np.pi, 100)

plt.figure()

plt.subplot(2, 2, 1)

plt.plot(X, np.sin(X))

plt.subplot(2, 2, 2)

plt.plot(X, np.cos(X))

plt.subplot(2, 1, 2)

plt.plot(X, np.tan(X))

plt.show()
```
